```
#Python program to create a function for salary bonus calculator
def calculate_bonus(salary, experience):
    if experience >= 10:
        bonus = 0.20
    elif experience >= 5:
        bonus = 0.10
    else:
        bonus = 0.05

    return salary + (salary * bonus)

n = int(input("Enter number of employees: "))

for i in range(n):
    salary = float(input("Enter salary: "))
    exp = int(input("Enter years of experience: "))

    total = calculate_bonus(salary, exp)
    print("Final Salary:", total)
```
