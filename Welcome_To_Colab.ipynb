```
#Python program to check the list of numbers
def classify_number(num):
    if num > 0:
        sign = "Positive"
    elif num < 0:
        sign = "Negative"
    else:
        sign = "Zero"

    if num != 0:
        if num % 2 == 0:
            parity = "Even"
        else:
            parity = "Odd"
    else:
        parity = "Neither Even Nor Odd"

    return sign, parity

numbers = list(map(int, input("Enter numbers separated by space: ").split()))

for num in numbers:
    sign, parity = classify_number(num)
    print(num, "→", sign, ",", parity)
```
