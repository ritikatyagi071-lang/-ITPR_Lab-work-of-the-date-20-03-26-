```
#Python program to create a function for palindrome checker
def is_palindrome(value):
    value = str(value)
    rev = ""

    for i in range(len(value)-1, -1, -1):
        rev += value[i]

    if value == rev:
        return True
    else:
        return False

data = input("Enter number/string: ")

if is_palindrome(data):
    print("Palindrome")
else:
    print("Not Palindrome")
```
